#### 2.14 Create a subset of mlb dataset for Year 2010

```python
yr2010.head()
```

#### 2.15 Sort Operations

```python
# sort row labels
yr2010.sort_index().head()
```

```python
# sort column labels
yr2010.sort_index(axis=1).head()
```

```python
# sort column values using order field; note: the order field returns a series
print(yr2010.Salary.sort_values(ascending=False).head())
print(type(yr2010['Salary']), type(yr2010.Salary))
print(type(yr2010[['Salary']]))
```

```python
# sort column values using sort_index method
sorted_yr2010 = yr2010.sort_values(ascending=False, by = ['Salary'])
```

```python
sorted_yr2010.head(20)
```

```python
yr2010.sort_values(ascending=[False,True], by =['Salary', 'Team']).head(20)
```

```python
 # Top 10 highest paid players
top10 = yr2010.Salary.sort_values(ascending=False).head(10)

top11 = yr2010['Salary'].sort_values(ascending=False).head(11)
# if you apply sort_values() function on DataFrame then by argument is must.(But on Series it is not)
```

```python
print(type(top10),type(top11))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

#plot highest paid

top10.plot(label='Salaries')
plt.xticks(rotation='vertical')
plt.legend()
```
